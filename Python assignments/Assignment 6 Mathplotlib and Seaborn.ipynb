```python
plt.figure(figsize=(15,18))
plt.title('Bar plot shows the population travelling to Arrival cities for vacation ')
sns.barplot(y=air_fares['Arrival_CITY'], x=air_fares['Arrival_avg_population'], hue = air_fares['VACATION'])
```

```python
plt.figure(figsize=(15,18))
plt.title('Bar plot shows the Population travelling for vacation from corresponding departure cities ')
sns.barplot(y=air_fares['Departure_CITY'], x=air_fares['Departure_avg_population'], hue = air_fares['VACATION'])
```

```python
plt.figure(figsize=(15,16))
plt.title('Departure cities and their corresponding average income in bar plot')
sns.barplot(x=air_fares['Departure_avg_INCOME'],y=air_fares['Departure_CITY'])
```

```python
plt.figure(figsize=(15,16))
plt.title('Departure cities and their corresponding average income in line plot')
sns.lineplot(x=air_fares['Departure_avg_INCOME'],y=air_fares['Arrival_avg_INCOME'])
```

```python
plt.figure(figsize=(15,16))
plt.title('Scatter plot depicts the population from and to the citiez travelling for vacation ')
sns.scatterplot(x=air_fares['Departure_CITY'] , y=air_fares['Arrival_CITY'], hue=air_fares['VACATION'])
```

```python
plt.figure(figsize=(15,16))
sns.lmplot(data = air_fares, x='Departure_avg_population', y='Arrival_avg_population', hue='VACATION')
plt.title('Departure / Arrival population in corresponding to vacation ')
```

```python
plt.figure(figsize=(15,16))
plt.title('Departure cities and their corresponding traveling distance')
sns.barplot(x=air_fares['DISTANCE'],y=air_fares['Arrival_CITY'])
```

```python
plt.figure(figsize=(15,16))
sns.lmplot(data = air_fares, x='DISTANCE', y='FARE')
plt.title('Scatter plot depicts the travelling distance and fare ')
```
